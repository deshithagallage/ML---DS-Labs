```python
import pandas as pd
import numpy as np

data = pd.read_csv('Perceptions towards eXplainable AI (XAI) (Responses) - Form Responses 1.csv')
print(data)
```

```python
# Get null value count of 'Occupation level'
print(data['Occupation level'].isnull().sum())
```

```python
# Get the unique values of 'Age'
print(data['Age'].nunique())
```

```python
df = data.copy()

# Correct answers dictionary
correct_answers = {
    'Which of the following mobile phone app uses AI in its operations': 'Email',
    'Can an AI tool determine if you have risk of Pneumonia by examining your chest x-rays?': 'Yes',
    'When you try to unlock your mobile phone using a pin, does an underline AI tool make that decision?': 'No',
    'Think about the advertisements you see when you are on the internet. Are those selected by AI tools?': 'Yes'
}

# Add a new column 'AI awareness' with default value 0
df['AI awareness'] = 0

# Iterate over rows and assign marks based on correct answers
for idx, row in df.iterrows():
    marks = 0
    for question, answer in correct_answers.items():
        if row[question] == answer:
            marks += 1
    df.at[idx, 'AI awareness'] = marks

# Print the DataFrame to verify changes
print(df.head())
```

```python
df.rename(columns={'Assume that an AI system makes decisions for you in your day-to-day activities. How important is it to know the reasons behind such decisions?': 'Perceived Importance of XAI'}, inplace=True)


from scipy.stats import chi2_contingency

categorical_columns = ['Gender', 'Age', 'Education level', 'Field of occupation', 'Occupation level', 'AI awareness']
for column in ['AI awareness']:
    contingency_table = pd.crosstab(df[column], df['Perceived Importance of XAI'])
    
    # Perform chi-square test
    chi2_stat, p_val, dof, expected = chi2_contingency(contingency_table)
    
    # Print results
    print(f"Chi-square test results for {column}:")
    print("Chi-square statistic:", chi2_stat)
    print("p-value:", p_val)
    print("\n")
    print(contingency_table)
```
