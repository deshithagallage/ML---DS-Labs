```python
import pandas as pd
import numpy as np

attendance = pd.read_csv('attendance.csv')
employees = pd.read_csv('employees.csv')
holidays = pd.read_csv('holidays.csv')
leaves = pd.read_csv('leaves.csv')
salary = pd.read_csv('salary.csv')
salary_dictionary = pd.read_csv('salary_dictionary.csv')
```

```python
filtered_salary = salary[(salary['Basic Salary_0'] != 0)]
mean_basic_salaries = filtered_salary.groupby('Employee_No')['Basic Salary_0'].mean().reset_index()
employees_with_salary = pd.merge(employees, mean_basic_salaries, on='Employee_No', how='left')

filtered_salary = salary[(salary['Net Salary'] != 0)]
mean_net_salaries = filtered_salary.groupby('Employee_No')['Net Salary'].mean().reset_index()
employees_with_salary = pd.merge(employees_with_salary, mean_net_salaries, on='Employee_No', how='left')

# Rename the column to 'basic_salary'
employees_with_salary.rename(columns={'Basic Salary_0': 'Basic Salary'}, inplace=True)

print(employees_with_salary.head())
```

```python
print(attendance['Employee_No'].nunique())
```

```python
attendance['date'] = pd.to_datetime(attendance['date'])
attendance_filtered = attendance[attendance['date'] <= '2023-12-31']

print(attendance_filtered)
```

```python
attendance_unique_values = attendance_filtered['Employee_No'].value_counts().reset_index()
attendance_unique_values.columns = ['Employee_No', 'Attendance']

print(attendance_unique_values)
```

```python
employees_with_attendance = pd.merge(employees_with_salary, attendance_unique_values, on='Employee_No', how='left')
employees_with_attendance.head()
```

```python
employees_with_attendance.to_csv('output.csv', index=False)
```
